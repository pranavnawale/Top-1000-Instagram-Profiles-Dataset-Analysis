```python
# Top20 followers
top20=insta.nlargest(20,"Followers")
top20
```

```python
# Barplot
plt.figure(figsize=(13,9))
fig2=plt.bar(top20.Category,top20.Followers,label='Followers in millions',width=0.8,color='purple')
plt.xticks(fontsize=12,rotation=90)
plt.ylabel("Followers in millions",fontsize=16)
plt.xlabel("Category",fontsize=16)
plt.title("Top 20 Category by Followers",fontsize=20)
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
# Showing the category using pie chart
types=insta.Category.value_counts()
types
```

```python
plt.pie(types.values,labels=types.index,labeldistance=1.2,autopct="%1.2f%%",shadow=True,startangle=60)
plt.title("Different Category",fontsize=20)
plt.show()
```
